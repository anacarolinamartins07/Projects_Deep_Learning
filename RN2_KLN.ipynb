```python
# função entrada: CV saida, vaga - saida: score

pontuacao = ['(', ')', ';', ':', '[', ']', ',']
stop_words = nltk.corpus.stopwords.words('portuguese')

def MatchCV(cv, vagas, limite = 5):
    '''
    cv: caminho de um arquivo PDF
    vaga: dataset de palavras-chave e pesos
    '''

    arquivoPDF = pdf.open(cv) #abrindo o arquivo
    primeira_pagina = arquivoPDF.pages[0]  #lê apenas a primeira página
    textoCRU = primeira_pagina.extract_text() #lendo o arquivo

    lista_de_palavras = nltk.tokenize.word_tokenize(textoCRU)  # transforma o texto cru em uma lista de termos
    lista_de_palavras = [palavra.lower() for palavra in lista_de_palavras]  # deixando tudo minusculo

    keywords = [palavra for palavra in lista_de_palavras if not palavra in stop_words and not palavra in pontuacao]  # tira as pontuacoes e stopwords
    textocv = " ".join(s for s in keywords)  # junta tudo em um texto só novamente.

    pesos = list(vagas['pesos']) #criando a lista dos pesoas
    palavras_chaves = list(vagas['palavras-chave']) #criando lista para palavras chaves

    cont = [textocv.count(pc) for pc in palavras_chaves]  # conta quantas vezes cada termo da vaga aparece no texto do cv

    def aux(x, limite): #treshoold
        return x if x <= limite else limite

    cont = [aux(i, limite) for i in cont]   # coloca o limite na contagem de palavras com a funcao aux

    pmax = np.sum(np.array(pesos) * limite) #padronizando entre 0 e 1

    score = ((np.array(cont) * pesos).sum()/pmax).round(4) #para calcular o escore com a pontuacao maxima pmax(a pontuacao maxima que alguem pode tomar )

    return score
```

```python
#para analisar varios curriculos de uma vez

lista_de_vagas = vagas

lista_de_curriculos = [
    "/content/Nayara.pdf",
    "/content/Rafael.pdf",
    "/content/Sabrina.pdf",
    "/content/Victor.pdf"] #importando curriculos
```

```python
lista_de_vagas[0]
```

```python
MatchCV(lista_de_curriculos[1], lista_de_vagas[0],3) #chamando funcao
```

```python
# Lista de Listas, cada lista interna é score de uma pessoa nas vagas
escores = [[MatchCV(cv, vaga) for vaga in lista_de_vagas] for cv in lista_de_curriculos] #pontuacao: cada curriculo para cada vaga
escores
```

```python
# filtro para pegar apenas os nomes das pessoas na lista de caminhos
nomes = [cv.split('/')[-1].split('.')[0] for cv in lista_de_curriculos]
nomes
```

```python
matchs = pd.DataFrame(escores, columns = nomevagas, index= nomes ) #criando um df com os scores, os nomes e as vagas
matchs
```
